```python
# Importing required libraries for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Load the Dataset
df = pd.read_csv("traffic.csv")
df.head()

#Clean the Data
df.drop_duplicates(inplace=True)

df.dropna(inplace=True)

df['date'] = pd.to_datetime(df['date'])


#Compute Users
users = df['linkid'].nunique()
print(users)

#Compute Sessions
sessions = len(df)
print(sessions)

#Bounce Rate
events_per_link = df.groupby('linkid')['event'].count()

single_event = events_per_link[events_per_link == 1].count()

bounce_rate = (single_event / len(events_per_link)) * 100

print(bounce_rate)

#Average Session Duration
session_time = df.groupby('linkid')['date'].agg(['min','max'])

session_time['duration'] = (session_time['max'] - session_time['min']).dt.seconds

avg_duration = session_time['duration'].mean()

print(avg_duration)

#Top Landing Pages

landing_pages = df['linkid'].value_counts().head(10)

landing_pages.plot(kind='bar')

plt.title("Top Landing Links")
plt.show()

#Top Countries

top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar')

plt.title("Traffic by Country")

plt.show()

#Top Artists

top_artists = df['artist'].value_counts().head(10)

top_artists.plot(kind='bar')

plt.title("Top Artists")

plt.show()

#User Journey
journey = df['event'].value_counts()

journey.plot(kind='bar')

plt.title("User Events")

plt.show()

```

**Dataset Description**

The dataset used in this analysis contains records of website interactions. Each row represents a user interaction with a specific link or music content. The dataset contains the following columns:

date: Indicates the timestamp when the interaction occurred.
country: Shows the country from which the user accessed the link.
city: Represents the city location of the user.
artist: Name of the artist associated with the track.
album: Name of the album.
track: Title of the track.
isrc: International Standard Recording Code used to uniquely identify the recording.
linkid: Unique identifier for the link being accessed.
This dataset helps in understanding which content receives more engagement and how users interact with different links.
